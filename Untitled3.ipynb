```python
import zipfile
import os

# Find the zip file automatically
zip_files = [f for f in os.listdir() if f.endswith('.zip')]
print("Found zip file:", zip_files)

# Extract it
with zipfile.ZipFile(zip_files[0], 'r') as zip_ref:
    zip_ref.extractall('marketing_data')

# Show what's inside
print("Files extracted:", os.listdir('marketing_data'))
```

```python
import pandas as pd

# Get the CSV filename automatically
csv_files = [f for f in os.listdir('marketing_data') if f.endswith('.csv')]
print("CSV file found:", csv_files)

df = pd.read_csv('marketing_data/' + csv_files[0])
print(df.shape)
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())

print("\nDuplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())

df['Date'] = pd.to_datetime(df['Date'])
print("\nData cleaned successfully!")
print(df.columns.tolist())
```

```python
channel_summary = df.groupby('Campaign_Type').agg({
    'Impressions': 'sum',
    'Clicks': 'sum',
    'Engagement_Score': 'mean'
}).reset_index()

channel_summary['CTR (%)'] = (channel_summary['Clicks'] / channel_summary['Impressions']) * 100
channel_summary = channel_summary.sort_values('CTR (%)', ascending=False)

print(channel_summary)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(14,5))

ax[0].bar(channel_summary['Campaign_Type'], channel_summary['Impressions'], color='skyblue', label='Impressions')
ax[0].bar(channel_summary['Campaign_Type'], channel_summary['Clicks'], color='orange', label='Clicks')
ax[0].set_title('Impressions vs Clicks by Channel')
ax[0].legend()
ax[0].tick_params(axis='x', rotation=45)

ax[1].bar(channel_summary['Campaign_Type'], channel_summary['CTR (%)'], color='green')
ax[1].set_title('Click-Through Rate (%) by Channel')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('funnel_chart.png')
plt.show()
```

```python
print("FINAL CHANNEL SUMMARY TABLE:")
print(channel_summary.to_string(index=False))

best_channel = channel_summary.iloc[0]['Campaign_Type']
print(f"\n🏆 Best performing channel by CTR: {best_channel}")
```
